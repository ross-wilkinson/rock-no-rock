## Statistical Analysis (Anova RM)

```
from statsmodels.stats.anova import AnovaRM

#perform the repeated measures ANOVA
print(AnovaRM(data=df, depvar='power', subject='subject', within=['condition'], aggregate_func='mean').fit())
```

## Multiple Comparisons


```
import statsmodels.stats.multicomp as mc

comp = mc.MultiComparison(df['power'], df['condition'])
post_hoc_res = comp.tukeyhsd()
post_hoc_res.summary()
```

## Plot comparison

```
post_hoc_res.plot_simultaneous(ylabel= "Condition", xlabel= "Power Difference")
```
